```python
import torch
from torch import nn, functional as F
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib notebook
```

```python
torch.manual_seed(4)

# create 2d grid of points
X = torch.tensor(np.mgrid[-10:10:.5, -10:10:.5].reshape(2, -1).T) * 1.
H = nn.ReLU()(nn.Linear(2, 2, bias=True, dtype=float)(X))
Y = nn.ReLU()(nn.Linear(2, 2, bias=True, dtype=float)(H))

activations = [X.detach().numpy(), H.detach().numpy(), Y.detach().numpy()]

fig, axs = plt.subplots(1, len(activations), figsize=(5*len(activations), 5))

Y = activations[-1]
for act, ax in zip(activations, axs):
    colors = np.ones((Y.shape[0], 4))
    colors[:, :2] = Y / np.abs(Y).max(axis=0, keepdims=True)
    ax.scatter(act[:, 0], act[:, 1], c=colors)
    
plt.show()
```
